```python
% load_ext autoreload
% autoreload 2
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'
import os

opj = os.path.join

# adaptive-wavelets modules

# evaluation
from matplotlib import gridspec
from eval_simulation import load_results
```

# load results

```python
waves = ["db5", "sym5", "coif2"]
# result path
path = opj(os.getcwd(), "results")
dirs = [wave + "_saliency_warmstart_seed=1" for wave in waves]
dics, results, models = load_results(dirs, waves, path)
```

```python
index = 0  # select which result to display
lamwaves = np.fromiter(dics[index]['lamL1wave'].values(), dtype=float)
lamattrs = np.fromiter(dics[index]['lamL1attr'].values(), dtype=float)
lamwaves = np.unique(lamwaves)
lamattrs = np.unique(lamattrs)
R = len(lamwaves)
C = len(lamattrs)

plt.figure(figsize=(C + 1, R + 1), dpi=300)
gs = gridspec.GridSpec(R, C,
                       wspace=0.0, hspace=0.0,
                       top=1. - 0.5 / (R + 1), bottom=0.5 / (R + 1),
                       left=0.5 / (C + 1), right=1 - 0.5 / (C + 1))

for r in range(R):
    for c in range(C):
        ax = plt.subplot(gs[r, c])
        ax.plot(dics[index]['x'][(r, c)], dics[index]['psi'][(r, c)])
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.tick_params(
            axis='both',
            which='both',
            bottom=False,
            top=False,
            left=False,
            right=False,
            labelbottom=False)
        if c == 0:
            plt.ylabel(str(lamwaves[r]))
        if r == 0:
            plt.title(str(lamattrs[c]))

plt.savefig('figures/sim_wave_vary_db5.pdf', bbox_inches='tight')
```
